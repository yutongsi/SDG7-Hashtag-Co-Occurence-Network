# Collect all hashtags (non-repetitive)

```python
# Define a function to extract and print all feasible hastags(ignore if repetitive) in alphabetical order
def extract_all_hashtags(df_data):
    hashtags_list = []
    for idx, data in df_data.iterrows():
        # extract from entities column 
        data = eval(data.entities) # data needs to be type: dictionary for entities column
    
        # extract hashtags for each tweet
        if "hashtags" not in data.keys():
            continue
        else:
            hashtags = []
            for tags in data['hashtags']: 
                # obtain the hastag in lowercase
                tags_lower = tags['tag'].lower()
            
                # deal with repeated hashtags
                if tags_lower in hashtags: 
                    continue
                else:  # otherwise save the tag
                    hashtags.append(tags_lower)
            # add the data
            hashtags_list.extend(hashtags)

    # contains all hashtags non-repeated
    hashtags_set_sorted = sorted(set(hashtags_list)) # repeated ones will not be returned in the set
    
    return hashtags_set_sorted
```

```python
# extract all hashtags
all_hashtags_list = extract_all_hashtags(df_data) # apply the function 

# print
for tag_idx, tag in enumerate(all_hashtags_list):
    print(f"Idx = {tag_idx}, hashtag = {tag}")
    
# # save as a dataframe
# tweet_hashtags_df = pd.DataFrame(all_hashtags_list)
# tweet_hashtags_df.to_csv('./sdg7_all_hashtags.csv',index=False)
```

# Print/check 4 situations for irrelevant tweets - Deal with error data

Case 1) Data dropped in these 3 situations
- 1.1 situation 1: entities do not contain the "key" named "hashtags"
- 1.2 situation 2: entities have "hashtags" without the hashtag of interest such as "sdg7".
- 1.3 situation 3: entities have non-readable hashtags

Case 2) Data have problems but need to be kept in the dataset
- 2.1 situation 1: "hashtags" in entities have repeated hashtags such as "sdg7, sdg13, sdg13", we need to keep the data in the corpus (save only once).

```python
def check_error_hashtags(idx, data, hashtag_of_interest, non_readable_hashtags):
    # extract from entities column
    data = eval(data.entities) # data needs to be dictionary type for the entities column
    
    if "hashtags" not in data.keys(): # case 1 situation 1) if entities do not contain the key named "hashtags"
        print(f"Drop index = {idx} for case 1 situation 1. Existing keys = {data.keys()}")
        print("===================================================================================")
        return False, idx
    else:
        # save tags
        hashtags = []
        for tags in data['hashtags']:
            # obtain the hastag in lowercase
            tags_lower = tags['tag'].lower()
            
            # check if it is non-readable
            if tags_lower in non_readable_hashtags: # case 1 situation 3) non-readable hashtags in the list
                print(f"Drop index = {idx} for case 1 situation 3. Non-readable hashtags = {tags_lower}")
                print("==========================================================================================")
                return False, idx
            
            # deal with repeated hashtags
            if tags_lower in hashtags:  # case 2 situation 1) ignore repeated hashtags if already collected per tweet
                print(f"Modify index = {idx} for case 2 situation 1. Repeated hashtags = {tags_lower}")
                print("===========================================================================================")
                continue
            else:  # otherwise save the tag
                hashtags.append(tags_lower)
                
        # check for hashtags of interest
        if not (hashtag_of_interest in hashtags):  # case 1 situation 2) 
            print(f"Drop index = {idx} for case 1 situation 2. No hashtag of interest = {hashtags}")
            print("==========================================================================================")
            return False, idx
        else:
            return hashtags, idx
```

```python
tweet_hashtags_list = []  # save the hashtags for each tweet (does not contain dropped ones)
hashtag_of_interest = "sdg7"  # hashtag of interest
non_readable_hashtags = ["„Éï„Ç©„É≠„Éê100","Áõ∏‰∫í„Éï„Ç©„É≠„Éº"]  # list of non-readable hashtags
drop_count = 0  # counter of the dropped tweets
drop_index_list = []  # index list of the dropped tweets

# loop the data frame
for idx, d in df_data.iterrows():
    
    # extract the hashtags for one tweet
    res, res_idx = check_error_hashtags(idx, d, hashtag_of_interest, non_readable_hashtags)
    
    if not res:  # drop the raw tweet if any conditions above have been met
        drop_count += 1  # increase the counter by 1
        drop_index_list.append(res_idx) # save the index of the dropped tweet
    else:        
        tweet_hashtags_list.append(res)  # save the tweet that does not meet any condition above

# print the information of the dropped tweets
print(f"Drop number of tweet = {drop_count}")
print(f"Dropped tweet indices = {drop_index_list}")
```

```python
# drop data
#note that this will also remove some hashtags (printed in the former cell) since those are in the tweets
df_data_drop = df_data.drop(drop_index_list)
```
