```python
# This will import the Twarc2 client and expansions class from twarc library and also the json library
from twarc import Twarc2, expansions
import json
import datetime
import pandas as pd
import json
import re
```

```python
# This is where you initialize the client with your own bearer token
client = Twarc2(bearer_token="...")

#please keep bearer_token confidential
```

```python
#Aim: we want to get tweets containing the hashtag of "sdg7" (not retweets or replies, written in English).

# Specify the start time in UTC for the time period you want replies from
start_time = datetime.datetime(2022, 1, 1, 0, 0, 0, 0, datetime.timezone.utc)

# Specify the end time in UTC for the time period you want Tweets from
end_time = datetime.datetime(2022, 7, 1, 0, 0, 0, 0, datetime.timezone.utc)

# This is where we specify our query, which aims at extracting tweets containing the hashtag of "energyjustice."
query = "(#sdg7) -is:retweet -is:reply lang:en"

# Note that "has:geo" is redundant since you already put "place_country: US."

# Name and path of the file where you want the Tweets written to
file_name = 'sdg7.txt'

# The search_all method call the full-archive search endpoint to get Tweets based on the query, start and end times
search_results = client.search_all(query=query, start_time=start_time, end_time=end_time, max_results=100)
```

```python
#DON'T RUN THIS MULTIPLE TIMES (there is a restriction on tweets that can be retrieved within a timeframe). 

# Twarc returns all Tweets for the criteria set above, so we page through the results
id=0
dict_tweet={}
for page in search_results:
    # The Twitter API v2 returns the Tweet information and the user, media etc.  separately
    # so we use expansions.flatten to get all the information in a single JSON
    result = expansions.flatten(page)
    #We will open the file and append one JSON object per new line
    #with open(file_name, 'a+') as filehandle:
            #for tweet in result:
            #filehandle.write('%s\n' % json.dumps(tweet))
    for tweet in result:
        id += 1
        #if id == 100:
            #break #modiao
        dict_tweet[id]=tweet
        
#         print(tweet.keys())
        
    #if id == 100:
        #break #modiao
        
df_data=pd.DataFrame.from_dict(dict_tweet,orient='index')        
```

```python
# dict_tweet
df_data
```

```python
#drop data if the texts are totally repetitive, keep the last tweet
df_data=df_data.drop_duplicates(['text'], keep='last')
df_data
```

```python
#drop data if entities is nan

df_data=df_data.dropna(subset = ['entities'])
```
